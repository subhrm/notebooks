### Hypothesis

Repeted softmax is same

```python
import numpy as np
```

```python
def softmax(arr):
    exps = np.exp(arr)
    return exps/np.sum(exps)
```

```python
D = 5
a = (np.random.random(D) - 0.5)*2
print(f"{a=}")
s=softmax(a)
print(f"{s=}")
s2=softmax(s)
print(f"{s2=}")
s3=softmax(s2)
print(f"{s3=}")
s4=softmax(s3)
print(f"{s4=}")
```
