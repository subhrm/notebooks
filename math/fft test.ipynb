```python
import numpy as np
import cv2
import skimage as ski
import matplotlib.pyplot as plt
import scipy as sp
```

```python
img_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/1/1e/Peace_Of_Nature_%28129019177%29.jpeg/1024px-Peace_Of_Nature_%28129019177%29.jpeg"
img = ski.io.imread(img_url)
print(f"{img.shape}")
```

```python
img = img / 255.0
img
```

```python
plt.imshow(img)
```

```python
fft2 = np.fft.fft2(img)
fft2_abs = np.abs(fft2)
fft2_norm = fft2_abs/np.max(fft2_abs)

# Set K
K = 0.5
top_k = np.percentile(fft2_abs, 100-K)
fft2[fft2_abs<top_k] = 0
rev_img = np.fft.ifft2(fft2)
plt.imshow(np.abs(rev_img))
```
