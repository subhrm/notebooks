```python
N = 20
results = []
for a in range(N):
    a3 = a*a*a
    for b in range(N):
        b3 = b*b*b
        for c in range(N):
            r = a3 + b3 + c*c*c - 3*a*b*c
            results.append(r)

results = sorted(list(set(results)))
print(len(results))
print(results[:100])
```
