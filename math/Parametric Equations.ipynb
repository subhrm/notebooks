```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# circle
t = np.linspace(-5.0,5.0, 1000)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

```python
# knot
t = np.linspace(-10.0,10.0, 1000)
x = 0.25*t*np.cos(t)
y =  0.25*t*np.sin(t)
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

```python
# slope
t = np.linspace(1,5.0, 1000)
x = np.sqrt(t)
y =  np.sqrt(t)*np.cos(t)
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

```python
# circle
t = np.linspace(-5.0,5.0, 1000)
x = 1 - np.sin(t)
y =  1 - np.cos(t)
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

```python
# try some diff functions
t = np.linspace(-10,10.0, 1000)
x = t
y = np.sin(t)/t
plt.figure(figsize=(6,6))
plt.plot(x,y)
```

### Butterfly

```python
def butterfly_curve(t):
    x = np.sin(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-(np.sin((t/12)))**5)
    y = np.cos(t)*(np.exp(np.cos(t))-2*np.cos(4*t)-(np.sin((t/12)))**5)

    return x, y

time = np.arange(0, 12 * np.pi, 0.01)
x,y = butterfly_curve(time)

plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title('Butterfly Curve')

plt.show()
```

```python
t = np.linspace(-5.0,5.0, 1000)
x = np.sin(t*0.35) + np.tan(t)
y = np.cos(t*0.25)
plt.plot(x,y)

```
