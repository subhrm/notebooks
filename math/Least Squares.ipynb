```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
N = 200
x = np.linspace(-2, 2, N)
a, b = np.random.random(2) + 0.5
y = a*x + b + np.random.random(N)/3.0
plt.plot(x,y)
```

```python
print(a,b)
```

```python
w1, w2 = np.random.random(2)
alpha = 0.5
for _ in range(10):
    y_hat = x*w1 + w2 
    loss = np.mean(np.square(y_hat - y))
    # print(f"Loss = {loss}")

    dl_dw1 = (1.0/(2.0*N))*np.dot( (y_hat - y) , x )
    w1 = w1 - alpha*dl_dw1

    dl_dw2 = (1.0/(2.0*N))*np.sum( y_hat - y )
    w2 = w2 - alpha*dl_dw2
    # print(f"New {w1} {w2}")

y_hat = x*w1 + w2 
loss = np.mean(np.square(y_hat - y))
print(f"Loss = {loss}")
print(f"New {w1} {w2}")
```
