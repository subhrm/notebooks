## Fibonacci

```python
import numpy as np
import plotly.express as px
```

```python
F = np.array([[1,1],[1,0]])
```

```python
F
```

```python
evals, evecs = np.linalg.eig(F)
```

```python
evecs @ evecs.T 
```

```python
s = np.array([[1,0]]).T
s
```

```python
F@s
```

```python
F@F@F@F@F@s
```

```python
evecs @ np.diag(evals) @ evecs.T
```

```python
# 100th Fibo number
```

```python
evecs @ np.diag(evals**10) @ evecs.T @ s
```

```python
F = F.astype(np.int64)
s = np.array([1,0]).reshape((2,1)).astype(np.int64)
fib_values = [ s.flatten() ]
for _ in range(60):
    s_new = F@s
    fib_values.append(s_new.flatten())
    s = s_new
```

```python
len(fib_values), fib_values[-1]
```

```python
px.scatter(x=[x[0] for x in fib_values], y=[x[1] for x in fib_values], trendline="ols")
```
