```python
import numpy as np
import plotly.express as px
```

```python
def square_root(a, eps=1e-6):
    x = 1
    err = 1
    values = [x]
    while(err > eps):
        x_next = 0.5*(x + (a/x) )
        err = abs(x_next - x)
        x = x_next
        values.append(x)

    return x, values
```

```python
N = 25
res, values = square_root(N, eps=1e-9)
print(res, len(values))
px.scatter(x=np.arange(len(values)), y=values)
```
