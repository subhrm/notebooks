```
fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

    => P = xA 

    (B-P).T * P = 0

    (B-xA).t * xA = 0

    (B-xA).t * A = 0

    B.T*A - xA.T*A = 0 
    B.T*A  = xA.T*A
    x = B.T*A / (A.T*A) = A.t*B / (A.t*A)

    P = xA = A*A.t*B / (A.t*A) 
    = (A*A.t) * B / (A.t*A) 


    = (A*A.t)/(A.t*a)  * B
