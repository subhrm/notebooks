```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = 256
N = 10000
cosine_similarities = []
for _ in range(N):  # Generate 1000 angles
    # Generate two random vectors around the origin
    vec1 = np.random.rand(D) - 0.5
    vec1 = vec1/ np.linalg.norm(vec1)
    vec2 = np.random.rand(D) - 0.5
    vec2 = vec2/np.linalg.norm(vec2)
    # Calculate the dot product
    dot_product = np.dot(vec1, vec2)
    cosine_similarities.append(dot_product)

print(f"{np.mean(cosine_similarities)=}")
print(f"{np.max(cosine_similarities)=}")
print(f"{np.min(cosine_similarities)=}")


from scipy.stats import norm
# Calculate mean and standard deviation of your data
mu, std = norm.fit(cosine_similarities)
# Generate x-values for the fitted distribution
xmin, xmax = np.min(cosine_similarities), np.max(cosine_similarities)
x = np.linspace(xmin, xmax, 100)
# Generate y-values for the fitted distribution
p = norm.pdf(x, mu, std)


# Plot the histogram
plt.hist(cosine_similarities, bins=100,density=True, alpha=0.6, color='g')
# plot the fitted distribution
plt.plot(x, p, 'k', linewidth=2)
plt.xlabel("cosine_similarities")
plt.ylabel("Frequency")
plt.title("Histogram of cosine_similarities Between pairs of Random Vectors")
plt.show()
```
