## Derivatves of dot procut
#### If  $y=x^T.x$
#### then $Δy = 2x^T.Δx$


```python
x = np.random.random(6)*10
print(f"{x=}")
y = x@x
print(f"{y=}")
dx = np.random.random(6)/200.0
print(f"{dx=}")
x_new = x + dx
dy_actual = x_new@x_new - y
print(f"{dy_actual=}")

dy_computed = 2*x@dx
print(f"{dy_computed=}")
```

# Derivative of a Matrix Square

### if $Y = X@X$
### then $\Delta Y = X@ \Delta X + \Delta X @ X$

```python
X = np.random.random((5,5))*10
print(f"{X=}")
Y = X@X
print(f"{Y=}")
dX = np.random.random((5,5))/20.0
print(f"{dX=}")
X_new = X + dX
dY_actual = X_new@X_new - Y
print(f"{dY_actual=}")

dY_computed = X@dX + dX@X
print(f"{dY_computed=}")

np.abs(dY_computed - dY_actual)
```
