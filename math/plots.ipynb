## Notebook for Plotting form functions

```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

### 2. Plot some functions

```python
# Create x values from -4*pi to 4*pi
x = np.linspace(-2.0*np.pi, 2.0*np.pi, 500)

# Calculate y values (sin(x))
y1 = np.sin(x)
y2 = np.sin(1.0/x)
y3 = x*np.sin(1.0/x)
y4 = x*x*np.sin(1.0/x)

# Create the figure object
fig = go.Figure()

# Add the traces using the figure's add_trace method
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='sin(x)'))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='sin(1/x)'))
fig.add_trace(go.Scatter(x=x, y=y3, mode='lines', name='x*sin(1/x)'))
fig.add_trace(go.Scatter(x=x, y=y4, mode='lines', name='(x^2)*sin(1/x)'))

# Show the figure
fig.show()
```

## Gausian Integral
## $\int_{-\infty}^{\infty} e^{-x^2} \,dx  = \sqrt$

```python
x = np.linspace(-10,10,10000)
# Calculate the Gaussian function values
y = np.exp(-1.*x*x)
# Calculate the width of each trapezoid for numerical integration
dx = (np.max(x) - np.min(x))/len(x)
# Perform numerical integration using the sum of areas of trapezoids
integral = np.sum(y)*dx
# The expected value of the Gaussian integral
exp = np.sqrt(np.pi)
# Print the calculated integral, expected value, and the absolute difference
print(f"{integral=:.6f} {exp=:.9f} {abs(integral-exp)=:.6f}")
# Plot the Gaussian function
px.line(x=x,y=y)
```
