```python
# import libs
import numpy as np
```

```python
# prompt: Generate a random 5 by 5 symmetric matrix and calculate it's eigen values

A = np.random.rand(5, 5)
A = (A + A.T) / 2 + 1.0*np.eye(5)
print(A)
print(np.linalg.eigvals(A))
```

```python
# prompt: Use QR factorization to repeatatively compute Q,R and the A = RQ
X = A
for _ in range(10):
    Q, R = np.linalg.qr(X)
    X = R@Q
    # X[X < 0.0000000001] = 0
print(np.diag(X))
```
