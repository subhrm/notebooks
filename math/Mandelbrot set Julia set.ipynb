```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def madelbort(c,n=10):
    z = c
    for _ in range(n):
        z = z*z + c
        # v_new = v*v + c
        # # print(v, v_new)
        # delta = abs(v_new) / abs(v)
        # v = v_new

    # print(delta)
    # return delta <= 1.1
    return abs(z) <= 2

print(f"{madelbort(0.17+0.2j)=}")
print(f"{madelbort(0.7+0.2j)=}")
```

```python
def complex_matrix(xmin, xmax, ymin, ymax, pixel_density):
    re = np.linspace(xmin, xmax, pixel_density)
    im = np.linspace(ymin, ymax, pixel_density)
    # return im[:, np.newaxis]
    return re[np.newaxis, :] + im[:, np.newaxis] * 1j
```

```python
c = complex_matrix(-2, 0.5, -1.2, 1.2 ,10000)
print(c.shape)
mask = madelbort(c)
print(mask.shape)
members = c[mask]
members.shape
```

```python
plt.figure(figsize=(12,12))
plt.scatter(members.real, members.imag, color="orange", marker=".", s=1)
plt.gca().set_aspect("equal")
plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
def madelbort2(c,n=10):
    z = c
    for _ in range(n):
        z = z*z + c
    return abs(z)

v = madelbort2(c)
print(v.shape)
v[v>2.0] = 2.0
v = v / 2.0
```
