```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk(steps=1):
    x,y = 0,0
    moves = [(1,0),(-1,0),(0,1),(0,-1)]
    result = [0]
    for _ in range(steps):
        dx,dy = random.choice(moves)
        x,y = x+dx, y+dy
        result.append((x*x + y*y)**0.5)

    return np.array(result)


print(random_walk(1))
```

```python
n_obs = 200
M = 10000

points = np.array([ [i for _ in range(n_obs)] for i in range(M+1)  ])
dist = np.vstack([ random_walk(M) for _ in range(n_obs)  ])

plt.figure(figsize=(24,10))
plt.tick_params(labeltop=True, labelright=True)
plt.scatter(points.flatten(), dist.T.flatten(),s=2 )
plt.xticks(ticks=[i for i in range(0,M+1,100)], rotation = 45)
plt.yticks(ticks=[i for i in range(0,int(np.max(dist))+1,10)])
plt.xlim([0, M])
plt.ylim([0, int(np.max(dist))+1])
plt.plot( np.arange(M+1), np.mean(dist, axis=0), c="yellow", linewidth=4 )
plt.grid(which="both", c="lightgreen")
```
