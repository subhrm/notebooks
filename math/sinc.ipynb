# Plot the Sinc Function

## $ \frac{Sin(x)}{x} $

```python
import numpy as np
import plotly.express as px
```

```python
x = np.linspace(-16*np.pi, 16*np.pi, 10000)
y = np.sin(x) / x

px.scatter(x=x,y=y)
```

# Integration

## $ \int \limits^ _{-} \frac{Sin(x)}{x} dx  =  $

## Finite integration using `numpy`

```python
# finite integration
N = 20000
M = 40
x = np.linspace(-M*np.pi, M*np.pi, N)
y = np.sin(x) / x
dx = (x[-1] - x[0]) / N
print(f"{dx=}")

integration = np.sum(y*dx)
print(f"{integration=} {np.pi=}")
```

## Using `scipy`

```python
import scipy as sp
```

```python
left_val , _ = sp.integrate.quad(lambda x: np.sin(x)/x, -40.0*np.pi ,0)
right_val , _ = sp.integrate.quad(lambda x: np.sin(x)/x, 0 , 40.0*np.pi)
print(f"{left_val=} {right_val=} {left_val  + right_val}")
```
