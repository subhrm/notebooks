```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
A = widgets.FloatSlider(value=1, min=-10.0, max=10.0, step=0.1,description='A:')
B = widgets.FloatSlider(value=1, min=-5.0, max=5.0, step=0.1,description='B:')
C = widgets.FloatSlider(value=1, min=-5.0, max=5.0, step=0.1,description='C:')
Bttn = widgets.Button(description="Draw")
x = np.linspace(-2,2,200)

def plot_figure(*args):
    clear_output()
    # display the widgets again
    display(widgets.HBox([A,B,C,Bttn]))

    # Get the values of sliders
    a,b,c =(w.value for w in (A,B,C))

    y = a*x*x + b*x + c 
 
    fig = px.line(x=x,y=y)

    # Solve for optimal point
    p = (-1.0*b)/(2.0*a)
    min_y = a*p*p + b*p +c 
    print(f"f(x) is minimal at x={p} with value: {min_y}")

    fig.add_trace(go.Scatter(x=x,y=np.ones_like(x)*min_y))
    fig.add_trace(go.Scatter(x=np.ones_like(y)*p,y=y))

    fig.show()


Bttn.on_click(plot_figure)
display(widgets.HBox([A,B,C, Bttn]))
```
