```python
import numpy as np
```

```python
a = np.ones(shape=(3,4,2))
a.shape
```

```python
# updating some values
a[0,0,0] = 10
a[0,0,1] = 20
a[0,1,0] = 30
a[0,1,1] = 40
```

```python
a[:,:,0]  # The first slice in X-Y Plane
```

```python
a[:,:,1]  # Second slice in X-Y Plane
```

### The `'C'` or `'F'` ordering of NumPy array is particularly relevant when NumPy arrays are used in interfaces with low level libraries written in C and Fortran.

```python
a.flags
```

#### `numpy.nditer` would iterate the `n-d array` the way it is stored

```python
" ".join( str(i) for i in np.nditer(a) )
```
