## Newton Raphson Method of approximation

$$X_{n+1} = X_n - \frac{F(n)}{F'(n)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Lets find $$\sqrt{n}$$

$$f(x) = x^2 - n$$
$$f'(x) = 2*x$$

```python
def sqrt(n):
    f = lambda x:x*x - n
    f_prime = lambda x: 2*x

    eps = 1e-7
    x = 3
    while(abs(f(x)) > eps):
        x_new = x - f(x)/f_prime(x)
        print(x,x_new)
        x = x_new

    return x
```
