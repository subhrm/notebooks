```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = np.array([[0.8, 0.1], [0.2, 0.9]])
```

```python
A
```

```python
def plt_vecs(*matrices):
    n = len(matrices)
    fig, axes = plt.subplots(nrows= 1, ncols=n, figsize=(8*n, 8))
    
    for i, mat in enumerate(matrices):
        ax = axes[i]
        for row in mat.T:
            ax.plot([0,row[0]], [0,row[1]])
            
```

```python
plt_vecs(A,A)
```

```python
B = np.array([[1,1, 3], [2,1, 4]])
plt_vecs(A, B, A@B)
```

```python
V = np.array([[-0.4472136 ],[-0.89442719]])
```

```python
plt_vecs(A,V, A@(V*-45))
```
