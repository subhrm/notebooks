```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
A = np.array([ [1,1], [1,2], [1,3], [1,4], [1,5]])
A
```

```python
b = np.array([3,4.2, 4.9, 5.85,7.1])
b
```

```python
plt.scatter(A[:,1], b)
```

```python
w = np.linalg.inv(A.T @ A) @ A.T @ b.reshape((5,1))
w
```

```python
model = LinearRegression()
model.fit(A,b)
model.coef_, model.intercept_
```

```python
plt.scatter(np.arange(1,6), b, label="orginal points")
plt.plot(np.arange(1,6), model.predict(A).flatten(), label="sklearn")
# plt.plot(np.arange(1,6), (A@w).flatten(), label="Linear ALgebra")
plt.legend()
```

### Fit a parabola

```python
x = np.linspace(0.1,3,50)
c = 0.5
w1 = 0.3
w2 = 1.75

y = c + w1*x + w2*x*x
plt.plot(x,y)
```

```python
A = np.hstack([np.ones(50).reshape((50,1)), x.reshape((50,1)), (x*x).reshape((50,1))])
A.shape
```

```python
b = y + np.random.normal(0,1.0,50)/6.0
```

```python
plt.plot(x,b)
```

```python
w = np.linalg.inv(A.T @ A) @ A.T @ b.reshape((-1,1))
w
```

```python
p = A@w
plt.plot(x,b)
plt.plot(x,p.flatten())
```
