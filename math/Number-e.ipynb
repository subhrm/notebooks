### Compute $e$ using standard formula
---

 $$ \lim_{N \to ∞} (1+\frac{1}{N})^N $$

```python
M = 60 
a = np.arange(1,M+1,1)
e_approx = np.power( 1.0 + (1.0/a), a)
plt.figure(figsize=(18,6))
_ = plt.plot(a,e_approx)
plt.xticks(np.arange(0,M+1,5))
plt.grid()
```

### Compute $e$ using Taylor Series Formula
---

 $$ \sum_{N = 0}^{∞} \frac{1}{N!} $$

```python
N = 20
factorials = [1,1]
for i in range(2,N):
    factorials.append(factorials[-1]*i)
print(len(factorials), ":" ,factorials)
```

```python
x = np.arange(N)
f = 1.0/ np.asarray(factorials, dtype=np.int64)
y = np.cumsum(f)
print(y)
plt.figure(figsize=(18,6))
plt.plot(x,y, label="Computed e")
plt.plot(x, np.ones_like(x)*np.e, label="Actual e")
plt.xlabel("Number of Terms in Taylor Series")
plt.xticks(x)
plt.yticks(np.arange(0,3.1, step=0.2))
plt.grid()
_=plt.legend(loc='lower right')
```
