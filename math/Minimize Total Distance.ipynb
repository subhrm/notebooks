# Find a Point with Minimize the total distnace 

* Given a set of points `S` find a point `x` such that, sum of distnace of all points in `S` to `x` is minimum.  

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
N = 20
p = np.random.randint(low=-10, high=11, size=(N,2))
c = np.mean(p, axis=0)
print("Center is at ", c)
distance_from_mean = np.linalg.norm(p-c, axis=1)
total_distance_from_mean = np.sum(distance_from_mean)
print(f"Total distnace from mean: {total_distance_from_mean}")

distance_from_or = np.linalg.norm(p, axis=1)
total_distance_from_or = np.sum(distance_from_or)
print(f"Total distnace from origin (0,0): {total_distance_from_or}")

plt.figure(figsize=(12,8))
plt.scatter(p[:,0], p[:,1])
_=plt.scatter(c[0], c[1], c="red", s=20, marker="D")
```
