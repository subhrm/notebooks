```python
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
import plotly.express as px
import plotly.graph_objects as go
```

```python
A = widgets.IntSlider(value=1, min=-10, max=10, step=1,description='A:')
B = widgets.IntSlider(value=1, min=-10, max=10, step=1,description='B:')
C = widgets.IntSlider(value=1, min=-10, max=10, step=1,description='C:')
D = widgets.IntSlider(value=1, min=-10, max=10, step=1,description='D:')
E = widgets.IntSlider(value=1, min=-10, max=10, step=1,description='E:')
F = widgets.IntSlider(value=1, min=-10, max=10, step=1,description='F:')

x = np.linspace(-2,2,200)
y = np.linspace(-2,2,200)

def plot(*args):
    clear_output()
    # display the widgets again
    display(widgets.HBox([A,B,C]))
    display(widgets.HBox([D,E,F]))

    # Get the values of sliders
    a,b,c,d,e,f=(w.value for w in (A,B,C,D,E,F))

    z = np.vstack([
        a*xi*xi + b*y*y + c*xi*y + d*xi + e*y + f
        for xi in x
    ])
    # print(z.shape)
    fig = go.Figure(data=[
        go.Surface(z=z, x= x, y=y)
    ])
    fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
    fig.update_layout(width=1200,height=600,
                      scene = {
            "xaxis": {"nticks": 20},
            "yaxis": {"nticks": 20},
            "zaxis": {"nticks": 20},
            "aspectratio": {"x": 2, "y": 2, "z": 1}
        })

    # Solve for optimal point
    M = np.array([[a,c/2],[c/2,b]])
    t = -0.5*np.array([d,e])
    
    print(M,t,np.linalg.det(M) )
    
    if (np.linalg.det(M) > 0) and np.all(np.diag(M) >= 0 ) :
        print("We have a solution")
        X = np.linalg.inv(M)@t
        print(X)
        x1,x2 = X
        fig.add_scatter3d(x=np.ones(10)*x1, y=np.ones(10)*x2, z=np.arange(10), mode='lines' )
    else:
        print("No solution")
        

    fig.show()


for w in (A,B,C,D,E,F):    w.observe(plot, names='value')

display(widgets.HBox([A,B,C]))
display(widgets.HBox([D,E,F]))
```
