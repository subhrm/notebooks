### QR Factorization using Gram-Schimdt Algorithm Implemented Using Numpy

```python
import numpy as np
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
```

### Plot a 2-D Matrix

```python
def plot_2d_matrix(A):
    # fig = ff.create_quiver(x=np.zeros(2), y=np.zeros(2), u=A[0], v=A[1])
    # fig.show()
    # plt.figure()
    origin = np.zeros((2,2))
    plt.quiver(*origin, A[0], A[1], scale=10)
    plt.xlim([-2,5])
    plt.ylim([-2,5])
plot_2d_matrix(np.array([[1,2],[0,2]]))
```
