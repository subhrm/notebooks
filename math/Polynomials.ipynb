```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
class MyPoly: 
    def __init__(self, degree=3):
        self.params = (np.random.random(degree+1) - 0.5)*4.0
        print(f"Parameters = {self.params=}")    

    def compute(self, a):
        res = np.zeros_like(a) + self.params[0]

        for i,parm in enumerate(self.params[1:]):
            res += np.power(a, i+1)*parm

        return res

```

```python
x = np.linspace(-10,10,800)
P = MyPoly(3)
y = P.compute(x)
px.line(x=x,y=y, ) 
# plt.plot(x,y)
```
