## Hypothesis: Two random vectors in higher dimmension is very likely to be perpendicular

Intersting fact, if we chosse from Normal distribution with mean=0 and Variance = 1/D, the length of the vecs would tend to 1

```python
import numpy as np
```

```python
D =768
results = []
for i in range(100):
    # Generate 2 random factors of size, D
    # and find their dot products
    f1 = np.random.rand(D) - 0.5
    f2 = np.random.rand(D) - 0.5
    f1 = f1/np.linalg.norm(f1)
    f2 = f2/np.linalg.norm(f2)
    dot_product = np.dot(f1,f2)
    # print(f"{i} # {dot_product= }")
    results.append(dot_product)

print(np.abs(np.mean(results)))
```

```python
# do as matrix op
N,D = 10,512 # N samples of D dim vectors
gen = np.random.default_rng(seed=2024)
mat = gen.normal(loc=0.0, scale=np.sqrt(1.0/D), size=(N,D))
# Normalize the vecs to make then Unit length
# mat = mat/np.linalg.norm(mat,axis=1,keepdims=True)
print(mat.shape)
dot_pod = (mat @ mat.T).round(decimals=2)
print(dot_pod.shape)
print(dot_pod)
```
