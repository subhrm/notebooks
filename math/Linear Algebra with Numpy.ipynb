### Check if a matrix is positive definite.

```python
def is_positive_definte(mat):

    if len(mat.shape) != 2:
        print(f"Need  2-D matrix, got: {mat.shape}")
        return False

    m,n = mat.shape
    if m != n:
        print(f"Need  square matrix, got: {mat.shape}")
        return False
    if not (mat.T == mat).all():
        print(f"Need  symmetric matrix")
        return False
    det = np.linalg.det(mat)
    if det < 0:
        print(f"Det {det} is less than zero")
        return False

    eig_values = np.linalg.eigvals(mat)
    if not (eig_values > 0).all():
        print(f"All eigen values are not positive {eig_values}")
        return False
    
    return True
```

```python
a = np.array([[3,4],[4,6]])
is_positive_definte(a)
```

```python
a = np.array([[-3,4],[4,-6]])
is_positive_definte(a)
```
