```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# prompt: create a 1x2 sublots and plot sin and cosine curve

fig, axes = plt.subplots(1, 2)

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

axes[0].plot(x, y_sin)
axes[0].set_title('Sin Curve')

axes[1].plot(x, y_cos)
axes[1].set_title('Cosine Curve')

plt.tight_layout()
plt.show()
```

```python
# prompt: # prompt: create a 1x2 sublots and plot sin and cosine curve only using plotly

from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig = make_subplots(rows=1, cols=2, subplot_titles=("Sin Curve", "Cosine Curve"))

fig.add_trace(go.Scatter(x=x, y=y_sin, mode='lines', name='Sin'), row=1, col=1)
fig.add_trace(go.Scatter(x=x, y=y_cos, mode='lines', name='Cos'), row=1, col=2)

fig.update_layout(title_text="Sin and Cosine Curves")
fig.show()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
import numpy as np
```

```python
X = np.random.random((10,3))
X[:,0] = X[:,0] * 4.0
X[:,1] = X[:,1] + 2.0
X[:,2] = X[:,2]*2.0 + 5.0
print(X)
```

```python
est = KBinsDiscretizer(
    n_bins=3, encode='ordinal', strategy='uniform'
)
est.fit(X)
Xt = est.transform(X)
print(f"{est.n_features_in_=}\n{est.n_bins_=} \n{est.bin_edges_=}")
Xt
```

```python
est = KBinsDiscretizer(
    n_bins=3, encode='onehot-dense', strategy='quantile'
)
Xt = est.fit_transform(X)
print(f"{est.n_features_in_=}\n{est.n_bins_=} \n{est.bin_edges_=}")
print(f"{Xt.shape=}")
```
