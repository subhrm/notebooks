```
import numpy as np
import scipy
import scipy.special
import matplotlib.pyplot as plt
%matplotlib inline
```

### Poisson Distribution

```
X = np.arange(0,10)
# scipy.special.factorial(X)
l = 4.8
Y = (np.power(l,X) / scipy.special.factorial(X))*np.exp(-1*l)
plt.plot(X,Y)
```

```
t = np.linspace(-5,5, 5000)
y = np.exp(-1*t*t)
plt.plot(t,y)
```
