```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
models = ["GBM", "MLP", "SVR"]
scores_1 = [91, 92, 87]
scores_2 = [54,68,70]
```

```
N = 3
width = 0.3
ind = np.arange(N)  # the x locations for the groups

fig, ax = plt.subplots()
rects1 = ax.bar(ind, scores_1, width, color='b')
rects2 = ax.bar(ind+width, scores_2, width, color='y')

ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('GBM',  'MLP', 'SVR' ))

legend = ax.legend((rects1[0], rects2[0]), ('0.003', '0.002'), loc='upper right', bbox_to_anchor=(0.5, 1.2))
# Put a nicer background color on the legend.
legend.get_frame().set_facecolor('C0')

ax.set_xlabel("Model Names")
ax.set_ylabel("score")
ax.grid(color='gray', linestyle='-', linewidth=0.5)
plt.show()
```
