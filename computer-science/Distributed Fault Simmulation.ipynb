```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline
```

```python
nodes = 500_000
users = int(nodes*0.2)
chunks = 100_000

node_fault_ratio = 0.05
bad_nodes = np.random.randint(0,nodes-1, int(nodes*node_fault_ratio)) 
print(bad_nodes.shape)
```

```python
bad_nodes = np.sort(bad_nodes)
```

```python
affected_users = 0
file_fault_cnt = 0
replication_factor = 7
step = 100
for i in tqdm(range(0,users,step)):
    dist = np.random.randint(0,nodes-1, int(chunks*replication_factor))
    is_bad_chunk = np.isin(dist, bad_nodes).reshape((chunks,replication_factor))
    is_entirely_bad = np.sum(is_bad_chunk, axis=-1 )
    is_entirely_bad[is_entirely_bad<replication_factor] = 0
    is_entirely_bad[is_entirely_bad==replication_factor] = 1
    bad_file = np.sum(is_entirely_bad)
    file_fault_cnt += bad_file
    if bad_file > 0:
        affected_users += 1

print()
print(f"Total number of bad files : {file_fault_cnt*step}")
print(f"Total number of Affected users: {affected_users*step}")
```

```python
ratio = (file_fault_cnt*step) / (users*chunks)
reliability = 100.0 - ratio*100
print(ratio*100 , reliability)
```

```python
print(f"%of users affected : {((affected_users*step)/users)*100.0}")
```
