```python
# prompt: Generate  ablobs dataset with scikit-learn and plot it. Then do PCA transformation and plot the result.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
# Generate a 2D blobs dataset with 2 clusters
X, y = make_blobs(n_samples=300, centers=2, n_features=2, random_state=0, cluster_std=0.25)

# Plot the original dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis")
plt.title("Original Dataset")
plt.show()
```

```python

# Apply PCA transformation
pca = PCA(n_components=2)
X_transformed = pca.fit_transform(X)

# Plot the transformed dataset
plt.figure(figsize=(8, 6))
plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y, cmap="viridis")
plt.title("Transformed Dataset after PCA")
plt.show()
```
