```python
import torch
import torch.nn as nn
```

```python
word_emb_dim = 48
input_seq_len = 128

rnn = nn.LSTM(input_size=word_emb_dim, hidden_size=12, num_layers=1)
batch_size =32
input = torch.randn(input_seq_len,batch_size, 48)
h0 = torch.randn(1, 32, 12)
c0 = torch.randn(1, 32, 12)
output, (hn, cn) = rnn(input, (h0, c0))
print(f"{output.shape =} {hn.shape =} {cn.shape =}")
```
