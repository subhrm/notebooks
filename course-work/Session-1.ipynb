## Make first character of each word in a string uppercase

Ex: 

- Input : hello world   Output : Hello World
- Input:  subhendu mishra   output: Subhendu Mishra

```python
def capitalize_each_word(s: str) -> str:
    """
    Capitalize each word in a given string

    1- Split the string to words
    2- Capitalize each word
    3- Join with spaces
    """
    words = s.split()
    capitalized = [ w.capitalize() for w in words ]
    return " ".join(capitalized)
```

```python
capitalize_each_word("subhendu ranjan     mishra")
```

```python
"subhendu ranjan     mishra hello world vfjfh ghfghfh".split()
```

```python
def test_capitalize_each_word():
    assert capitalize_each_word("subhendu mishra") == "Subhendu Mishra" 
    assert capitalize_each_word("hello world") == "Hello World"
    assert capitalize_each_word("hello world") == "Hello orld"  # would fail
```

### List comprehension

Creating another list from a given list, with some operation over each element of the input list

```python
# ex 1: Sum of square of each number till 10
a = list(range(1,11)) 
b = [ i*i for i in a ]
res = sum(b)
print(a,b,res)
```

```python
# ex 1: Sum of square of each number till 100, if the number is divisible by 7 
a = list(range(7,101,7)) 
b = [ i*i for i in a ]
res = sum(b)
print(a,b,res)
```

```python
# ex 1: Sum of square of each number till 100, if the number is divisible by 7
# Print properly
a = list(range(7,101,7)) 
b = [ i*i for i in a ]
res = sum(b)
for x,y in zip(a,b):
    print(x, " --> ", y)

print(sum(a), "-->", sum(b))
```

```python
def digit_to_char(n):
    n = int(n)
    if n >= 10:
        raise Exception("Not a digit")
    
    text = [ "zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
    return text[n]


digit_to_char(7)   
```

```python
##  Convert any given number to indivigial digits in character form 

# 1947 - one nine four seven

def num_to_char(n):

    return " ".join([ digit_to_char(a) for a in str(n) ])


num_to_char(1947)
```
