```
events = [(11,100,120),(23,101,105), (34,106,125),(46,108,117)]
el =[]
for i,s,e in events:
    el.append((1,i,s))
    el.append((2,i,e))

print(el)

#sort el
el.sort(key=lambda x:x[0])
el.sort(key=lambda x:x[2])

print("Sorted el:")
print(el)

cnt,res=0,0
for cat,i, t in el:
    if cat ==1:
        cnt += 1
    else:
        cnt -= 1

    res = max(res,cnt)

print("Max # of concurrent  events = " , res)
```
