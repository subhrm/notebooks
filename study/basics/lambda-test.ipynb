```python
def add_to(num):
    def f(num2):
        return num + num2

    return f

add_to_ten = add_to(10)
print(add_to_ten(3)) # Returns 13
```

```python
def add_to_new(num):
    f = lambda num2: num + num2
    return f

add_to_five = add_to_new(5)
print(add_to_five(3)) # Returns 8
```
