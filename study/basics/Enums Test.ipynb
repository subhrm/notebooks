```python
from enum import Enum, auto
```

```python
class WorkspacePermission(Enum):
    CAN_DELETE_WORKSPACE = auto()
    CAN_ADD_USERS = auto()
    CAN_REMOVE_USERS = auto()
    CAN_UPLOAD_DOCUMENTS = auto()
    CAN_DELETE_DOCUMENTS = auto()
    CAN_CREATE_CONVERSATIONS = auto()

```
