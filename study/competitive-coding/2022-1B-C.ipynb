```python
import random
```

```python

def get_bin(n) -> str:
    op = bin(n)[2:]
    if len(op) < 8:
        op = "0" * (8 - len(op)) + op

    return op
```

```python
b = random.randint(1,255)
# b = 41
print(b, get_bin(b))

for i in range(300):
    bb = sum((int(c) for c in get_bin(b)))
    if bb < 4:
        x = 255
    else:
        r = random.randint(0,7)
        v = "00000001"
        w = v[r:] + v[:r]
        x = int(w,2)
    b = b^x
    print(i, x, get_bin(x), b, get_bin(b))
    if b == 255:
        print("Yay!!!")
        break
```
