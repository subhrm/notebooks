```python
import string
import os
import sys
```

```python
chars = string.ascii_lowercase
chars
```

```python
char_to_int = {c:i for i,c in enumerate(chars)}
# char_to_int
```

```python
raw_message = "Qixnu Nufupe Qovi Kufe Xufenus Bogynyce Woviqyze Getepugo Guze Deqy Hoxevux Funwi Zujizy Qoxy Woro Pudo Jyryrux Zysiky Qufi Tegqe Pozebes Xityfud Xofovul Levylo Pinji Jidycid Zicuwij Vulilin Wuxuwodu"
raw_message_words = list(raw_message.lower().split())
print(raw_message_words)
```

```python
for idx in range(26):
    new_words = []
    for rmw in raw_message_words:
        new_word = ""
        for c in rmw:
            #print(idx, rmw, c)
            ci = char_to_int[c]
            ci = (ci+idx) % 26
            new_word += chars[ci]
        new_words.append(new_word)

    print(f"{idx:2d}  decrypted message: {' '.join(new_words)}")
```
