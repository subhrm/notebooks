```python
import os
import sys
import requests
from bs4 import BeautifulSoup
import pandas as pd
import arrow
```

```python
URL = "https://www.petroldieselprice.com/Delhi-diesel-price"
```

```python
resp = requests.get(URL)
```

```python
resp.status_code
```

```python
doc = BeautifulSoup(resp.content, 'html.parser')
# doc
```

```python
all_tables = doc.find_all("table", class_="shop_table")
print(len(all_tables))
```

```python
all_tables[1].find_all("tr")
```

```python
def extract_from_table(table):
    date=[]
    rate=[]
    all_rows = table.find_all("tr")
    for row in all_rows[1:]:
        d,r,_ = row.find_all("td")
        date.append(d.string)
        rate.append(float(r.string[2:]))

    return date,rate

print(extract_from_table(all_tables[1]))
```
