```python
# for every stock get their last 6-months historical data using yfinace
COL_TO_USE = "Close"
N_OBS = 75

historical_data = []
long_names = []
industries = []

for stock in tqdm(top_stocks):
  try:
    yf_data = yf.Ticker(stock + ".NS")
    info = yf_data.info

    # Get stock name
    stock_name = info.get("longName", info.get("shortName" , stock))
    long_names.append(stock_name)
    # get the industry
    indistry = info.get("industry", "Unknown")
    industries.append(indistry)

    hist = yf_data.history(period="6mo")
    hist_data = hist[COL_TO_USE][-N_OBS:]
    # normalize and store
    historical_data.append( (hist_data - np.mean(hist_data)) / np.std(hist_data) )
  except Exception as e:
    print(f"Error downloading data for {stock}: {e}")

data = np.vstack(historical_data)
```

```python
print(f"{data.shape=}")
```

### Test data

```python
from numpy.testing import assert_almost_equal
# assert that the mean of all rows are almost zero
assert_almost_equal(np.mean(data, axis=1), np.zeros(data.shape[0]), decimal=2)
# assert that the std of all rows are almost one
assert_almost_equal(np.std(data, axis=1), np.ones(data.shape[0]), decimal=2)
```

## Plot data

```python
# reduce to 2-d using TSNE and plot using plotly

tsne = TSNE(n_components=2, random_state=42, perplexity=5)
reduced_data = tsne.fit_transform(data)

df = pd.DataFrame(reduced_data, columns=['x', 'y'])
df["stock"] = top_stocks
df["long_name"] = long_names
df["industry"] = industries

fig = px.scatter(df, x="x", y="y", color="industry", hover_data=["stock", "long_name","industry" ])
fig.show()
```
