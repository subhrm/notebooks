```python
import numpy as np
import skimage as ski
import scipy as sp
import matplotlib.pyplot as plt
```

```python
img = ski.io.imread("https://upload.wikimedia.org/wikipedia/commons/thumb/0/07/Fuel_truck_tanzania.jpg/1280px-Fuel_truck_tanzania.jpg")
print(img.shape)
ski.io.imshow(img)
```

```python
img_gray = ski.color.rgb2gray(img)
print(img_gray.shape)
ski.io.imshow(img_gray)
```

```python
np.max(img_gray), np.min(img_gray)
```

```python
mean_kernel = np.ones((3,3))/9.0
print(mean_kernel)
```

```python
smooth = sp.signal.convolve2d(img_gray, mean_kernel, mode='same')
smooth[smooth>1.0] = 1.0
print(smooth.shape, np.max(smooth), np.min(smooth))
ski.io.imshow(smooth)
```

### Try basic Prewitt Filters

```python
p_v_kernel = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
p_h_kernel = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
img_p_v = sp.signal.convolve2d(img_gray, p_v_kernel, mode='same')
img_p_h = sp.signal.convolve2d(img_gray, p_h_kernel, mode='same')

img_p_v[img_p_v>1.0] = 1.0
img_p_h[img_p_h>1.0] = 1.0

print(img_p_v.shape, img_p_h.shape)
fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(img_gray, cmap='gray')
ax[1].imshow(img_p_v, cmap='gray')
ax[2].imshow(img_p_h, cmap='gray')

# set titles
ax[0].set_title('Original')
ax[1].set_title('Prewitt Vertical')
ax[2].set_title('Prewitt Horizontal')
```
