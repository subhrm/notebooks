# TF IDF Vectorizer from scratch


```python
import numpy
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import re
```

```python
token_pattern = r"(?u)\b\w\w+\b"
text = "hello how are you? Are you okay? Yes, I am fine."
words = re.findall(token_pattern, text)
print(words)
```
