```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 4000
A,B = 0.75, 0.35
x = np.linspace(-5,5, N)
y = A*x + B
plt.plot(x,y)
```

```python
reg = LinearRegression().fit(x.reshape(-1, 1), y)
reg.score(x.reshape(-1, 1), y)
```

```python
reg.coef_ , reg.intercept_
```

### Let's add some normal noise

```python
noise = np.random.normal(scale=0.25, size=N) 
y_noise = y + noise
reg_noise = LinearRegression().fit(x.reshape(-1, 1), y_noise)
print("Score " , reg_noise.score(x.reshape(-1, 1), y_noise))
print(reg_noise.coef_ , reg_noise.intercept_)
plt.plot(x,y_noise)
a,b = reg_noise.coef_[0] , reg_noise.intercept_
y_pred = a*x + b 
plt.plot(x,y_pred,color="orange")
```

## Add some random noise

```python
noise = np.random.random(N)/5.0
y_noise = y + noise
reg_noise = LinearRegression().fit(x.reshape(-1, 1), y_noise)
print("Score " , reg_noise.score(x.reshape(-1, 1), y_noise))
print(reg_noise.coef_ , reg_noise.intercept_)
plt.plot(x,y_noise)
a,b = reg_noise.coef_[0] , reg_noise.intercept_
y_pred = a*x + b 
plt.plot(x,y_pred,color="orange")
```
