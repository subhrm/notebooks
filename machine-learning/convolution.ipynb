```
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```
a = np.arange(49,dtype=np.float).reshape((7,7,1))
```

```
print(a[0,0], a[0,1] )
```

```
data = np.expand_dims(a, axis=0)
data.shape
```

```
mask = np.array( [ [1,0,1],[0,2,0], [1,0,1]], dtype=np.float)
print(mask)
```

```
mask_new = np.expand_dims(np.expand_dims(mask,-1),-1)
mask_new.shape
```

```
res = tf.nn.conv2d(data ,mask_new,strides=1,padding='VALID').numpy()
res.shape
```
