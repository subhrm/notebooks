```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
N  = 10000
sample_size  = int(N*0.1)
C  = 100
selected_count = Counter()
```

```python
for _ in range(C):
    sample = np.random.randint(0, high=N, size=sample_size)
    selected_count.update(sample)
```
