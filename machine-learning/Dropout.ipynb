PyTorch Droput scales outputs. 

```python
import torch  # pip install torch
import torch.nn as nn
import numpy as np

'''nn.Dropout expects Float '''
test = torch.FloatTensor([1,0,5,2,0,3,6])
# test = np.array([1.0,0,5.0,2.0,0,3.0,6.0])
print(test)  # returns: tensor([1., 0., 5., 2., 0., 3., 6.])

for i in range(2):
    test_with_dropout = nn.Dropout(p=0.25)(test)  # random dropout
    print(test, test_with_dropout)
```

Tensorflow Dropout does not !

```python
# Try with Tensorflow 
import numpy as np
import tensorflow as tf
test = np.array([1.0,0,5.0,2.0,0,3.0,6.0])
for i in range(2):
    tf_dropout_output = tf.keras.layers.Dropout(rate=0.25)(test)  # random dropout
    print(test, tf_dropout_output)
```
