## Now lets try `unmasker` pipeline

```python
unmasker = pipeline("fill-mask")
```

```python
mask = unmasker.tokenizer.mask_token
print(f"Mask : {mask}")
```

```python
from pprint import pprint
```

```python
res = unmasker(f"Let go to a {mask}. I am very very hungry.")
pprint(res[:5])
```

```python
res = unmasker(f"It allows creating {mask} that can run inside a container.")
pprint(res[:5])
```
